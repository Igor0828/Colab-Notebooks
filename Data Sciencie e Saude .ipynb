```python
import pandas as pd
```

```python
confirmados = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
```

```python
confirmados.head(10)
```

```python
confirmados.sample(5)
```

```python
confirmados.set_index("Country/Region").loc["China"].sum().tail()
```

```python
china = confirmados["Country/Region"] == "China"
china
```

```python
confirmados["china"] = china
confirmados.set_index("Country/Region").loc["China"]
```

```python
confirmados.query("china==True")
```

```python
confirmados.groupby("china").sum()
```

```python
confirmados.groupby("china").sum() ['3/2/20'].plot(kind = 'pie')
```

```python
confirmados.groupby("china").sum() ['3/9/23'].plot(kind = 'bar')
```

```python
import numpy as np
```

```python
last_date = "3/9/23"

confirmados[last_date]
```

```python
confirmados['China_label']= np.where(confirmados["china"], 'China', 'Outros')
```

```python
confirmados
```

```python
confirmados.groupby("China_label").sum()[last_date].plot(kind='bar')

```

```python
import matplotlib.pyplot as plt

soma = confirmados.groupby("China_label").sum()
soma["delta"] = soma["3/9/23"] - soma["3/8/23"]
soma["delta"].plot(kind = 'bar')
plt.title(f"Novos Casos em {last_date}")
plt.show()
```

```python
soma.T.head()
```

```python
soma.T[2:-2].plot()
plt.title("Desenvolvimento de casos confirmados com o pasar do tempo")
```

```python
diferencas = soma.T[2:-2].diff()
```

```python
diferencas.plot()
plt.title("Casos novos com o decorrer do tempo")
plt.show()
```

```python
cases = confirmados.groupby("China_label").sum()[last_date].loc["China"]
estimativa_populacao = 1425671352
incidencia_ratio = cases / estimativa_populacao * 100000
incidencia_ratio
```
